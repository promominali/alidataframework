# aliframework API client CRUD-style demo

```python
from aliframework.api import ApiClient
from aliframework.config import ApiConfig, ApiAuthType

cfg = ApiConfig(base_url='https://httpbin.org', auth_type=ApiAuthType.NONE)
client = ApiClient(cfg)

# CREATE (POST)
print('POST /post:', client.post('/post', json={'name': 'foo'}).status_code)
# READ (GET)
print('GET /get:', client.get('/get', params={'id': 1}).status_code)
# UPDATE (PUT)
print('PUT /put:', client.request('PUT', '/put', json={'id': 1, 'name': 'bar'}).status_code)
# DELETE (DELETE)
print('DELETE /delete:', client.request('DELETE', '/delete', params={'id': 1}).status_code)
```
