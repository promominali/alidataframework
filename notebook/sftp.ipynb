# aliframework SFTP CRUD demo (files)

```python
from pathlib import Path
from aliframework.config import SftpConfig, SftpAuthType
from aliframework.sftp import create_sftp_client

# Write a local file to upload
Path('local.txt').write_text('hello from notebook')

cfg = SftpConfig(
    host='localhost',
    port=2222,
    username='user',
    password='secret',
    auth_type=SftpAuthType.PASSWORD,
)
sftp = create_sftp_client(cfg)

# CREATE/UPLOAD
sftp.put('local.txt', 'upload/remote.txt')
# READ/LIST
print('remote entries:', sftp.listdir('upload'))
# DOWNLOAD
sftp.get('upload/remote.txt', 'downloaded.txt')
# DELETE
sftp.remove('upload/remote.txt')
sftp.close()
```
